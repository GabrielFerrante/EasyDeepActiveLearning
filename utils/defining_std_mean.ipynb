```python
from data_utils import SVHNCustomDataset, calculate_dataset_stats
from torch.utils.data import DataLoader
from torchvision import transforms

# 1. Transformação básica apenas para análise
temp_transform = transforms.Compose([
    transforms.Resize((640, 640)),
    transforms.ToTensor() # Transforma para o intervalo [0, 1]
])

# 2. Carregar apenas o treino (ou treino + extra)
analysis_ds = SVHNCustomDataset('C:/Repositorios/EasyDeepActiveLearning/csv_SVHN/train.csv', 'F:/SVHN/train/train/', transform=temp_transform)
analysis_loader = DataLoader(analysis_ds, batch_size=16, shuffle=False, num_workers=4)

# 3. Calcular
mean, std = calculate_dataset_stats(analysis_loader)

print(f"\nResultados para o SVHN:")
print(f"Média (mean): {mean.tolist()}")
print(f"Desvio Padrão (std): {std.tolist()}")
```
